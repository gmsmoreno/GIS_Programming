## Add marker

```python
folium.Marker([-15.791528, -47.883040],
              popup = '<i>Shopping Conjunto Nacional Brasília</i>',
              tooltip = 'Click here!',
              icon = folium.Icon(color = 'red')
              ).add_to(map)
```

```python
map
```

## Create a Circle Buffer Marker

```python
folium.CircleMarker(location = [-15.791528, -47.883040],
                    radius = 50,
                    color = '#3186cc',
                    fill = True,
                    fill_color = '#3186cc').add_to(map)
```

```python
map
```

## Display lat/long coordinates on clicked map

```python
map.add_child(folium.LatLngPopup())
```

## Pandas Library

```python
import pandas as pd
from folium import plugins
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/olist_geolocation_dataset.csv')
```

```python
df.head()
```

```python
df = df.sample(frac = 0.03)
```

```python
coordinates = []
for lat, lng in zip(df.geolocation_lat, df.geolocation_lng):
  coordinates.append([lat, lng])
```

```python
map = folium.Map(location = [-15.791528, -47.883040], zoom_start = 13, tiles = 'Stamen Toner')
```

```python
map.add_child(plugins.HeatMap(coordinates))
map.add_child(folium.LatLngPopup())
```
